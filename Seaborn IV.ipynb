# Seaborn IV - Estilos y colores

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
propinas = sns.load_dataset('tips')
propinas.head()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=propinas)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=propinas)
sns.despine()
```

```python
sns.set_style('white')
sns.countplot(x='sex',data=propinas)
sns.despine()
```

```python
sns.set_style('white')
sns.countplot(x='sex',data=propinas)
sns.despine(left=True,bottom=True)
```

```python
plt.figure(figsize=(4,3))
sns.set_style('white')
sns.countplot(x='sex',data=propinas)
sns.despine()
```

```python
sns.set_context('notebook' ,font_scale=1.3)
plt.figure(figsize=(4,3))
sns.set_style('white')
sns.countplot(x='sex',data=propinas)
sns.despine()
```

# Obtener paleta de colores de seaborn para los cuadros

```python
sns.countplot(x='sex',data=propinas,palette='Set2')
```

```python
sns.countplot(x='sex',data=propinas,palette='BuGn_r')
```

```python
sns.countplot(x='sex',data=propinas,palette='PuBuGn_d')
```

```python
# Podemos utilizar colores https://htmlcolorcodes.com/
colores = ["#F5581E","blue"]
sns.countplot(x='sex',data=propinas,palette=colores)
```
